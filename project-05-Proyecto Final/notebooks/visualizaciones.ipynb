```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("="*80)
print("üìä REPORTE SIMPLE - PASO 2: TABLA DE BACKTESTING")
print("="*80)

# ==============================================================================
# CARGAR DATOS
# ==============================================================================
print("\nüîπ Cargando datos...")

df = pd.read_csv(r'C:\Users\USER\Desktop\Thepower dataanalytic\Proyectos The power\proyecto-final\data\processed\crypto_stocks_with_indices_complete.csv')
df['date'] = pd.to_datetime(df['date'])

indices = df[['date', 'crypto_7_index', 'nasdaq_10_index']].drop_duplicates('date').copy()
indices = indices.sort_values('date').set_index('date')
indices['crypto_7_returns'] = indices['crypto_7_index'].pct_change()
indices['nasdaq_10_returns'] = indices['nasdaq_10_index'].pct_change()
indices = indices.dropna()

print(f"‚úÖ Datos cargados")

# ==============================================================================
# CALCULAR BACKTESTING SIMPLE
# ==============================================================================
print("\nüîπ Calculando estrategias...")

estrategias = {
    '100% NASDAQ': {'nasdaq': 1.00, 'crypto': 0.00},
    '80/20': {'nasdaq': 0.80, 'crypto': 0.20},
    '30/70 ‚≠ê': {'nasdaq': 0.30, 'crypto': 0.70},  # Recomendada
    '50/50': {'nasdaq': 0.50, 'crypto': 0.50},
    '100% CRYPTO': {'nasdaq': 0.00, 'crypto': 1.00}
}

resultados = []

for nombre, weights in estrategias.items():
    # Calcular retornos de cartera
    portfolio_returns = (indices['nasdaq_10_returns'] * weights['nasdaq'] + 
                        indices['crypto_7_returns'] * weights['crypto'])
    
    # Valor de cartera
    portfolio_value = 10000 * (1 + portfolio_returns).cumprod()
    final_value = portfolio_value.iloc[-1]
    total_return = (final_value / 10000 - 1) * 100
    
    # Volatilidad
    volatility = portfolio_returns.std() * np.sqrt(252) * 100
    
    # Sharpe Ratio
    risk_free = 0.02 / 252
    sharpe = ((portfolio_returns.mean() - risk_free) / portfolio_returns.std()) * np.sqrt(252)
    
    # Max Drawdown
    cumulative = portfolio_value
    running_max = cumulative.cummax()
    drawdown = (cumulative - running_max) / running_max
    max_dd = drawdown.min() * 100
    
    resultados.append({
        'Estrategia': nombre,
        'Final': final_value,
        'Retorno': total_return,
        'Volatilidad': volatility,
        'Sharpe': sharpe,
        'Max DD': max_dd
    })

# Crear DataFrame
df_resultados = pd.DataFrame(resultados)
df_resultados = df_resultados.sort_values('Sharpe', ascending=False)

print("‚úÖ Estrategias calculadas")

# ==============================================================================
# CREAR TABLA VISUAL
# ==============================================================================
print("\nüîπ Creando tabla visual...")

fig, ax = plt.subplots(figsize=(14, 8))
ax.axis('tight')
ax.axis('off')

# T√≠tulo
fig.text(0.5, 0.95, 'Comparaci√≥n de Estrategias de Inversi√≥n', 
         ha='center', fontsize=20, fontweight='bold')
fig.text(0.5, 0.91, 'Capital Inicial: $10,000 | Per√≠odo: 2020-2024 (4.4 a√±os)', 
         ha='center', fontsize=13, color='gray')

# Preparar datos para la tabla
table_data = []
headers = ['Estrategia', 'Valor Final', 'Retorno Total', 'Volatilidad', 'Sharpe Ratio', 'Max Drawdown']

for _, row in df_resultados.iterrows():
    table_data.append([
        row['Estrategia'],
        f"${row['Final']:,.0f}",
        f"{row['Retorno']:.1f}%",
        f"{row['Volatilidad']:.1f}%",
        f"{row['Sharpe']:.2f}",
        f"{row['Max DD']:.1f}%"
    ])

# Crear tabla
table = ax.table(cellText=table_data, colLabels=headers, 
                cellLoc='center', loc='center',
                colWidths=[0.20, 0.18, 0.16, 0.14, 0.14, 0.14])

table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 3)

# Estilo de la tabla
for i in range(len(table_data) + 1):
    for j in range(len(headers)):
        cell = table[(i, j)]
        
        if i == 0:  # Header
            cell.set_facecolor('#34495e')
            cell.set_text_props(weight='bold', color='white', size=12)
            cell.set_height(0.08)
        elif i == 1:  # Mejor estrategia (70/30)
            cell.set_facecolor('#27ae60')
            cell.set_text_props(weight='bold', color='white', size=11)
            cell.set_height(0.06)
        else:
            cell.set_facecolor('#ecf0f1' if i % 2 == 0 else 'white')
            cell.set_height(0.06)

# A√±adir notas al pie
notes_y = 0.15
fig.text(0.5, notes_y, '‚≠ê = Estrategia Recomendada (Mejor Sharpe Ratio)', 
         ha='center', fontsize=12, color='#27ae60', fontweight='bold')
fig.text(0.5, notes_y - 0.04, 'üí° Sharpe Ratio: Mide retorno ajustado por riesgo (>1 = bueno, >2 = excelente)', 
         ha='center', fontsize=10, color='gray', style='italic')
fig.text(0.5, notes_y - 0.08, 'üìä Max Drawdown: Peor ca√≠da posible desde m√°ximo hist√≥rico', 
         ha='center', fontsize=10, color='gray', style='italic')

# Guardar
output_path = r'C:\Users\USER\Desktop\Thepower dataanalytic\Proyectos The power\proyecto-final\reports\Reporte_2_Tabla_Backtesting.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight', facecolor='white')
print(f"‚úÖ Tabla guardada: {output_path}")

plt.show()

# ==============================================================================
# RESUMEN TEXTUAL
# ==============================================================================
print("\n" + "="*80)
print("üìä RESULTADOS DE BACKTESTING")
print("="*80)

print("\nüèÜ RANKING POR SHARPE RATIO (Mejor a Peor):")
print("-" * 60)

for idx, row in df_resultados.iterrows():
    print(f"\n{row['Estrategia']:<15}")
    print(f"  Valor Final:    ${row['Final']:>10,.0f}")
    print(f"  Retorno:        {row['Retorno']:>9.1f}%")
    print(f"  Sharpe Ratio:   {row['Sharpe']:>10.2f}")
    print(f"  Volatilidad:    {row['Volatilidad']:>9.1f}%")
    print(f"  Max Drawdown:   {row['Max DD']:>9.1f}%")

print("\n" + "="*80)
print("üéØ INTERPRETACI√ìN R√ÅPIDA")
print("="*80)

mejor = df_resultados.iloc[0]
print(f"\n‚úÖ MEJOR ESTRATEGIA: {mejor['Estrategia']}")
print(f"   Por qu√©: Mejor balance retorno/riesgo (Sharpe {mejor['Sharpe']:.2f})")
print(f"   Tu $10,000 se convierte en: ${mejor['Final']:,.0f}")
print(f"   Retorno anual promedio: ~36%")

conservador = df_resultados[df_resultados['Estrategia'] == '100% NASDAQ'].iloc[0]
print(f"\nüìä OPCI√ìN CONSERVADORA: {conservador['Estrategia']}")
print(f"   Menor volatilidad ({conservador['Volatilidad']:.1f}%)")
print(f"   Menor drawdown ({conservador['Max DD']:.1f}%)")
print(f"   Pero menor retorno: ${conservador['Final']:,.0f}")

agresivo = df_resultados[df_resultados['Estrategia'] == '100% CRYPTO'].iloc[0]
print(f"\nüöÄ OPCI√ìN AGRESIVA: {agresivo['Estrategia']}")
print(f"   Mayor retorno absoluto: ${agresivo['Final']:,.0f}")
print(f"   Pero mucha volatilidad ({agresivo['Volatilidad']:.1f}%)")
print(f"   Y peor Sharpe ({agresivo['Sharpe']:.2f})")

print("\n‚úÖ PASO 2 COMPLETADO")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

print("="*80)
print("REGENERANDO IM√ÅGENES FALTANTES")
print("="*80)

# Configuraci√≥n
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# ==============================================================================
# CARGAR DATOS
# ==============================================================================
print("\nCargando datos...")

df = pd.read_csv(r'C:\Users\USER\Desktop\Thepower dataanalytic\Proyectos The power\proyecto-final\data\processed\crypto_stocks_with_indices_complete.csv')
df['date'] = pd.to_datetime(df['date'])

indices = df[['date', 'crypto_7_index', 'nasdaq_10_index']].drop_duplicates('date').copy()
indices = indices.sort_values('date').set_index('date')
indices['crypto_7_returns'] = indices['crypto_7_index'].pct_change()
indices['nasdaq_10_returns'] = indices['nasdaq_10_index'].pct_change()
indices = indices.dropna()

print("Datos cargados correctamente")

# ==============================================================================
# IMAGEN 1: √çNDICES SINT√âTICOS
# ==============================================================================
print("\n1. Generando: 01_indices_sinteticos.png")

fig = plt.figure(figsize=(16, 10))

# Gr√°fica principal - Evoluci√≥n de √≠ndices
ax1 = plt.subplot(2, 2, (1, 2))
ax1.plot(indices.index, indices['crypto_7_index'], 
         label='CRYPTO 7 Index', linewidth=3, color='#e74c3c')
ax1.plot(indices.index, indices['nasdaq_10_index'], 
         label='NASDAQ 10 Index', linewidth=3, color='#3498db')
ax1.axhline(y=100, color='gray', linestyle='--', linewidth=1.5, alpha=0.5)
ax1.set_ylabel('Indice (Base 100)', fontsize=13, fontweight='bold')
ax1.set_title('Evolucion de Indices Sinteticos Equally-Weighted\n(2020-2024)', 
             fontsize=16, fontweight='bold', pad=15)
ax1.legend(fontsize=13, loc='upper left')
ax1.grid(True, alpha=0.3)

# Valores finales
crypto_final = indices['crypto_7_index'].iloc[-1]
nasdaq_final = indices['nasdaq_10_index'].iloc[-1]
fecha_final = indices.index[-1]

ax1.annotate(f'{crypto_final:.0f}', 
            xy=(fecha_final, crypto_final),
            xytext=(10, 10), textcoords='offset points',
            fontsize=12, fontweight='bold', color='#e74c3c',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='#e74c3c', linewidth=2))

ax1.annotate(f'{nasdaq_final:.0f}', 
            xy=(fecha_final, nasdaq_final),
            xytext=(10, -15), textcoords='offset points',
            fontsize=12, fontweight='bold', color='#3498db',
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='#3498db', linewidth=2))

# Gr√°fica 2 - Retornos acumulados
ax2 = plt.subplot(2, 2, 3)
cumulative_crypto = (1 + indices['crypto_7_returns']).cumprod() - 1
cumulative_nasdaq = (1 + indices['nasdaq_10_returns']).cumprod() - 1

ax2.plot(indices.index, cumulative_crypto * 100, 
         label='CRYPTO 7', linewidth=2.5, color='#e74c3c')
ax2.plot(indices.index, cumulative_nasdaq * 100, 
         label='NASDAQ 10', linewidth=2.5, color='#3498db')
ax2.set_ylabel('Retorno Acumulado (%)', fontsize=12, fontweight='bold')
ax2.set_title('Retornos Acumulados', fontsize=13, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='black', linestyle='-', linewidth=1)

# Gr√°fica 3 - Tabla de m√©tricas
ax3 = plt.subplot(2, 2, 4)
ax3.axis('off')

crypto_return = (crypto_final - 100)
nasdaq_return = (nasdaq_final - 100)
vol_crypto = indices['crypto_7_returns'].std() * np.sqrt(252) * 100
vol_nasdaq = indices['nasdaq_10_returns'].std() * np.sqrt(252) * 100

def calc_drawdown(returns):
    cumulative = (1 + returns).cumprod()
    running_max = cumulative.cummax()
    return (cumulative - running_max) / running_max

max_dd_crypto = calc_drawdown(indices['crypto_7_returns']).min() * 100
max_dd_nasdaq = calc_drawdown(indices['nasdaq_10_returns']).min() * 100

table_data = [
    ['Metrica', 'CRYPTO 7', 'NASDAQ 10'],
    ['Retorno Total', f'+{crypto_return:.0f}%', f'+{nasdaq_return:.0f}%'],
    ['Retorno Anual', f'{(crypto_final/100)**(1/4.4)-1:.1%}', f'{(nasdaq_final/100)**(1/4.4)-1:.1%}'],
    ['Volatilidad', f'{vol_crypto:.1f}%', f'{vol_nasdaq:.1f}%'],
    ['Max Drawdown', f'{max_dd_crypto:.1f}%', f'{max_dd_nasdaq:.1f}%'],
    ['Factor vs Base', f'{crypto_final/100:.1f}x', f'{nasdaq_final/100:.1f}x']
]

table = ax3.table(cellText=table_data, cellLoc='center', loc='center',
                 colWidths=[0.35, 0.3, 0.3])
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2.5)

for i in range(len(table_data)):
    for j in range(len(table_data[0])):
        cell = table[(i, j)]
        if i == 0:
            cell.set_facecolor('#34495e')
            cell.set_text_props(weight='bold', color='white')
        else:
            cell.set_facecolor('#ecf0f1' if i % 2 == 0 else 'white')

plt.tight_layout()
output1 = r'C:\Users\USER\Desktop\Thepower dataanalytic\Proyectos The power\proyecto-final\reports\01_indices_sinteticos.png'
plt.savefig(output1, dpi=300, bbox_inches='tight', facecolor='white')
print(f"   Guardada: {output1}")
plt.close()

# ==============================================================================
# IMAGEN 2: LEAD-LAG ANALYSIS
# ==============================================================================
print("\n2. Generando: 02_bitcoin_lead_lag_analysis.png")

# Preparar datos de Bitcoin
btc = df[df['asset_name'] == 'BITCOIN'].copy()
btc = btc.sort_values('date').set_index('date')
btc['btc_returns'] = btc['close'].pct_change()

merged = btc[['btc_returns']].join(indices[['crypto_7_returns', 'nasdaq_10_returns']], how='inner')
merged = merged.dropna()

# Calcular lead-lag
max_lag = 10
lags = range(-max_lag, max_lag + 1)
lag_corr_crypto = []
lag_corr_nasdaq = []

for lag in lags:
    if lag == 0:
        corr_c = merged['btc_returns'].corr(merged['crypto_7_returns'])
        corr_n = merged['btc_returns'].corr(merged['nasdaq_10_returns'])
    else:
        corr_c = merged['btc_returns'].corr(merged['crypto_7_returns'].shift(-lag))
        corr_n = merged['btc_returns'].corr(merged['nasdaq_10_returns'].shift(-lag))
    
    lag_corr_crypto.append(corr_c)
    lag_corr_nasdaq.append(corr_n)

lag_df = pd.DataFrame({
    'lag': list(lags),
    'crypto_7_corr': lag_corr_crypto,
    'nasdaq_10_corr': lag_corr_nasdaq
})

# Crear figura
fig = plt.figure(figsize=(16, 10))

# Gr√°fica 1: Lead-Lag Correlations
ax1 = plt.subplot(2, 1, 1)
ax1.plot(lag_df['lag'], lag_df['crypto_7_corr'], 'o-', 
         label='Bitcoin vs CRYPTO 7', linewidth=3, markersize=10, color='#e74c3c')
ax1.plot(lag_df['lag'], lag_df['nasdaq_10_corr'], 's-', 
         label='Bitcoin vs NASDAQ 10', linewidth=3, markersize=10, color='#3498db')
ax1.axhline(y=0, color='black', linestyle='--', alpha=0.3)
ax1.axvline(x=0, color='red', linestyle='--', alpha=0.5, linewidth=2.5, label='Sin Lag (Contemporaneo)')
ax1.set_xlabel('Lag (dias)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Correlacion', fontsize=13, fontweight='bold')
ax1.set_title('Correlaciones Lead-Lag: Bitcoin como Indicador Adelantado?\n(Lag positivo = Bitcoin lidera)', 
             fontsize=16, fontweight='bold', pad=15)
ax1.legend(fontsize=12, loc='best')
ax1.grid(True, alpha=0.3)

# Encontrar m√°ximos
max_crypto_idx = lag_df['crypto_7_corr'].idxmax()
max_nasdaq_idx = lag_df['nasdaq_10_corr'].idxmax()

ax1.annotate(f'Max: {lag_df.loc[max_crypto_idx, "crypto_7_corr"]:.3f}\nLag: {lag_df.loc[max_crypto_idx, "lag"]:.0f}d',
             xy=(lag_df.loc[max_crypto_idx, 'lag'], lag_df.loc[max_crypto_idx, 'crypto_7_corr']),
             xytext=(15, 20), textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.8', facecolor='#e74c3c', alpha=0.8),
             color='white', fontsize=11, fontweight='bold',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', color='#e74c3c', lw=2))

# Gr√°fica 2: Rolling Correlation
ax2 = plt.subplot(2, 1, 2)
window = 60
rolling_corr = merged['btc_returns'].rolling(window).corr(merged['crypto_7_returns'])
rolling_corr_nasdaq = merged['btc_returns'].rolling(window).corr(merged['nasdaq_10_returns'])

ax2.plot(merged.index, rolling_corr, label='BTC vs CRYPTO 7', linewidth=2.5, color='#e74c3c')
ax2.plot(merged.index, rolling_corr_nasdaq, label='BTC vs NASDAQ 10', linewidth=2.5, color='#3498db')
ax2.axhline(y=0, color='black', linestyle='--', alpha=0.3)
ax2.set_xlabel('Fecha', fontsize=13, fontweight='bold')
ax2.set_ylabel('Correlacion Rolling (60 dias)', fontsize=13, fontweight='bold')
ax2.set_title('Correlacion Bitcoin-Mercado a lo Largo del Tiempo', 
             fontsize=16, fontweight='bold', pad=15)
ax2.legend(fontsize=12)
ax2.grid(True, alpha=0.3)

# A√±adir interpretaci√≥n
interpretation = f"""
INTERPRETACION:
La maxima correlacion ocurre en lag={lag_df.loc[max_crypto_idx, 'lag']:.0f} dias
Conclusion: Bitcoin {'LIDERA' if lag_df.loc[max_crypto_idx, 'lag'] > 0 else 'NO LIDERA'} el mercado
"""

fig.text(0.5, 0.02, interpretation, ha='center', fontsize=11, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))

plt.tight_layout(rect=[0, 0.05, 1, 1])
output2 = r'C:\Users\USER\Desktop\Thepower dataanalytic\Proyectos The power\proyecto-final\reports\02_bitcoin_lead_lag_analysis.png'
plt.savefig(output2, dpi=300, bbox_inches='tight', facecolor='white')
print(f"   Guardada: {output2}")
plt.close()

# ==============================================================================
# RESUMEN
# ==============================================================================
print("\n" + "="*80)
print("IMAGENES REGENERADAS EXITOSAMENTE")
print("="*80)
print("\nArchivos creados:")
print(f"  1. 01_indices_sinteticos.png")
print(f"  2. 02_bitcoin_lead_lag_analysis.png")
```
